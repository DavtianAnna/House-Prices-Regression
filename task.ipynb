# SGDRegressor + Pipeline + GridSearch

```python
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import  train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
model = SGDRegressor()
scale = StandardScaler()
```

```python
pipe = Pipeline([('scale', scale), ('model', model)])
```

```python
X_train = df_train.drop('SalePrice', axis=1)
y_train = df_train['SalePrice']
X_test = df_test.copy()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
params = {
    'model__penalty': ['l1', 'l2', 'elasticnet'],
    'model__alpha': [0.001, 0.01, 0.1, 1],
    'model__eta0': [0.001, 0.01, 0.1, 1],
    'model__max_iter': [100, 500, 1000]
}
```

```python
final_model = GridSearchCV(pipe, param_grid=params, verbose=2, cv=3, n_jobs = -1)
```

```python
final_model.fit(X_train, y_train)
```

```python
y_pred = final_model.predict(X_val)
```

```python
np.sqrt(mean_squared_error(y_pred, y_val))
```

```python
final_model.best_score_
```

```python
df_sub = pd.read_csv('sample_submission.csv')
finaly_y = final_model.predict(X_test)
df_sub['SalePrice'] = finaly_y
df_sub.to_csv('submission1.csv', index=False, index_label=False)
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import  train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
X_train = df_train.drop('SalePrice', axis = 1)
X_test = df_test.copy()
y_train = df_train['SalePrice']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor()
```

```python
params = {
    'n_estimators': [100, 150, 200],
    'max_depth': [4, 6, 8, 10, None]
}
```

```python
final_model = GridSearchCV(estimator=model, param_grid=params,cv=3, verbose=2)
```

```python
final_model.fit(X_train, y_train)
```

```python
y_pred = final_model.predict(X_val)
```

```python
np.sqrt(mean_squared_error(y_pred, y_val))
```

```python
df_sub = pd.read_csv('sample_submission.csv')
finaly_y = final_model.predict(X_test)
df_sub['SalePrice'] = finaly_y
df_sub.to_csv('submission2.csv', index=False, index_label=False)
```
